```python
# import pandas
import pandas as pd
# read csv
df = pd.read_csv('data/dafeatures.csv')
# Convert the DataFrame to a Dictionary
# data_dict = data.to_dict(orient='records')
df.head()
```

```python
nan_counts = df['preview_url'].isna().groupby(df['genre']).sum()

print("NaN counts in 'url' column split by 'genre':\n", nan_counts)
```

```python
nan_counts = df['preview_url'].notna().groupby(df['genre']).sum()

print("Actual counts in 'url' column split by 'genre':\n", nan_counts)
```
